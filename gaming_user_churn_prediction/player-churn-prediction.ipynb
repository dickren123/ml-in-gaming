我们尝试使用梯度增强树来模拟这个问题。SageMaker提供了一个XGBoost容器，我们可以使用它在托管的分布式环境中进行训练，然后作为实时预测端点托管。XGBoost使用梯度增强树，它自然地解释了特征和目标变量之间的非线性关系，同时也适应了特征之间的复杂交互。

Amazon SageMaker XGBoost可以对CSV或LibSVM格式的数据进行训练。对于本例，我们将继续使用CSV。为了适配算法训练的需要，数据准备为：

- 预测变量在第一列

- 没有标题行

首先把分类特征转换成数字特征。

````python
model_data = pd.get_dummies(churn)
model_data = pd.concat(
    [model_data["Churn?_True."], model_data.drop(["Churn?_False.", "Churn?_True."], axis=1)], axis=1
)
````

````python
model_data
````

数据集划分

````python
train_data, validation_data, test_data = np.split(
    model_data.sample(frac=1, random_state=1729),
    [int(0.7 * len(model_data)), int(0.9 * len(model_data))],
)
train_data.to_csv("train.csv", header=False, index=False)
validation_data.to_csv("validation.csv", header=False, index=False)
````

数据集上传到s3，供后续模型训练使用

````python
boto3.Session().resource("s3").Bucket(bucket).Object(
    os.path.join(prefix, "train/train.csv")
).upload_file("train.csv")
boto3.Session().resource("s3").Bucket(bucket).Object(
    os.path.join(prefix, "validation/validation.csv")
).upload_file("validation.csv")
````

### 模型训练

直接使用SageMaker内置的XGBoost算法，这里指定算法所在的容器镜像

````python
container = sagemaker.image_uris.retrieve("xgboost", boto3.Session().region_name, "1")
display(container)
````

设置s3中存放的数据信息

````python
s3_input_train = TrainingInput(
    s3_data="s3://{}/{}/train".format(bucket, prefix), content_type="csv"
)
s3_input_validation = TrainingInput(
    s3_data="s3://{}/{}/validation/".format(bucket, prefix), content_type="csv"
)
````

配置超参并启动训练任务

更多信息请见 [XGBoost](https://github.com/dmlc/xgboost/blob/master/doc/parameter.md).

````python
sess = sagemaker.Session()

xgb = sagemaker.estimator.Estimator(
    container,
    role,
    instance_count=1,
    instance_type="ml.m4.xlarge",
    output_path="s3://{}/{}/output".format(bucket, prefix),
    sagemaker_session=sess,
)
xgb.set_hyperparameters(
    max_depth=5,
    eta=0.2,
    gamma=4,
    min_child_weight=6,
    subsample=0.8,
    silent=0,
    objective="binary:logistic",
    num_round=100,
)

xgb.fit({"train": s3_input_train, "validation": s3_input_validation})
````

---
### 模型推理部署

我们使用SageMaker Endpoint来快速构建高可用、自动伸缩、自动负载均衡、A/B测试的模型线上服务

````python
xgb_predictor = xgb.deploy(
    initial_instance_count=1, instance_type="ml.m4.xlarge", serializer=CSVSerializer()
)
````

### 模型评估

现在已经运行了一个托管端点，我们通过在以下环节构建客户端发送http post请求来模拟线上的用户流失预测场景。

但首先，我们需要设置序列化程序和反序列化程序，来实现数据推理的前处理和后处理

这里定义一个方法来

1. 循环测试数据集

2. 将其拆分为小批量行

3. 将这些小批量转换为CSV字符串有效负载

4. 通过调用XGBoost端点检索小批量预测

5. 收集预测并将模型提供的CSV输出转换为NumPy数组

````python
def predict(data, rows=500):
    split_array = np.array_split(data, int(data.shape[0] / float(rows) + 1))
    predictions = ""
    for array in split_array:
        predictions = ",".join([predictions, xgb_predictor.predict(array).decode("utf-8")])

    return np.fromstring(predictions[1:], sep=",")

predictions = predict(test_data.to_numpy()[:, 1:])
````

我们通过一个简单的混淆矩阵来验证模型的有效性

````python
pd.crosstab(
    index=test_data.iloc[:, 0],
    columns=np.round(predictions),
    rownames=["actual"],
    colnames=["predictions"],
)
````

在247个流失玩家中，我们已经正确预测了其中的235个（True Positive）。而且，我们错误地预测有12个玩家会流失，但最终没有这样做（False Positive）。也有12个客户最终流失，模型没有预测到（False Negative）。

但是这里有一个参考阈值的概念。

````python
plt.hist(predictions)
plt.show()
````

如果我们手动调整阈值

````python
pd.crosstab(index=test_data.iloc[:, 0], columns=np.where(predictions > 0.3, 1, 0))
````

### 相对误差成本计算

由于阈值的存在，会导致我们采取不同运营手段来挽回和关怀玩家。FP和FN都会给我们带来额外的成本浪费

比如，大家可以根据自己实际的运营成本来设计一个玩家流失成本计算公式：

```txt
500 * FN(C) + 0 * TN(C) + 100 * FP(C) + 100 * TP(C)
```

````python
cutoffs = np.arange(0.01, 1, 0.01)
costs = []
for c in cutoffs:
    costs.append(
        np.sum(
            np.sum(
                np.array([[0, 100], [500, 100]])
                * pd.crosstab(index=test_data.iloc[:, 0], columns=np.where(predictions > c, 1, 0))
            )
        )
    )

costs = np.array(costs)
plt.plot(cutoffs, costs)
plt.show()
````

````python
print(
    "支出成本最小的阈值是:",
    cutoffs[np.argmin(costs)],
    "成本预估:",
    np.min(costs),
)
````
